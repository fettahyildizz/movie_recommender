```
import numpy as np
import pandas as pd
```

```
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data',sep='\t', names = column_names)
df.head()
```

```
movie = pd.read_csv('Movie_Id_Titles')
df = pd.merge(df,movie)
df.head()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('white')
%matplotlib inline
```

```
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```
ratings['num of ratings'] = df.groupby('title')['rating'].count()
ratings.head()
```

```
plt.figure(figsize=(10,4))
ratings['num of ratings'].hist(bins=70)
```

```
ratings['rating'].hist(bins=70)
```

```
sns.jointplot(y = 'num of ratings',x='rating',data=ratings)
```

```
moviemat = df.pivot_table(index='user_id', columns='title', values='rating')
moviemat.head()
```

Top 10 most rated movie:

```
ratings.sort_values(by=['num of ratings'], ascending=False).head(10)
```

```
sw = moviemat['Star Wars (1977)']
liar_liar = moviemat['Liar Liar (1997)']
sw.head()
```

We use corrwith() method to get correlations between two series.

```
similiar_to_sw = moviemat.corrwith(sw)
similiar_to_liar = moviemat.corrwith(liar_liar)
```

We drop NaN values. If we sort the values, we get some results that don't make any sense. This is because there are a lot of movies only watched once by users who also watched Star Wars.

```
corr_sw = pd.DataFrame(similiar_to_sw, columns=['correlation'])
corr_sw.dropna(inplace=True)
corr_sw.sort_values(by='correlation', ascending=False).head(5)
```

Let's filter out the movies which have less than 100 reviews.

```
corr_sw = corr_sw.join(ratings['num of ratings'])

corr_sw = corr_sw[corr_sw['num of ratings']>=100]
corr_sw.head()
```

Let's sort the rows by correlation values.

```
corr_sw.sort_values('correlation', ascending=False).head()
```

```
corr_liar = pd.DataFrame(similiar_to_liar, columns=['correlation'])
corr_liar['num of ratings'] = ratings['num of ratings']
corr_liar = corr_liar[corr_liar['num of ratings']>100]
corr_liar.sort_values('correlation', ascending=False).head()
```
